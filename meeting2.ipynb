## Website 1: Olympic Medal Statistics   

Imports and getSoup function:

```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests

def getSoup(url: str) -> BeautifulSoup:
    # your implementation
    page = requests.get(url)
    bs = BeautifulSoup(page.content,"html.parser")
    page.close()
    
    return bs
```

Website URL:
(hint: you need to create a soup object)

```python

url = "https://en.wikipedia.org/wiki/All-time_Olympic_Games_medal_table"
soup = getSoup(url)
print(soup.prettify())
# get soup object
```

#### Exercise 1.1: Getting the data
 Scrape the first table and put it into a pandas df with columns 'Country', 'Gold', 'Silver', 'Bronze', and 'Total'

```python
# Scrape and put into df here
tables = soup.find("table", class_="wikitable sortable")

df = pd.DataFrame(columns=['Country', 'Gold', 'Silver', 'Bronze', 'Total'])
for row in tables.tbody.find_all('tr'):
    columns = row.find_all('td')
    
    if(columns != []):
        country = columns[0].text.strip()
        gold = columns[2].text.strip()
        silver = columns[3].text.strip()
        bronze = columns[4].text.strip()
        total = columns[5].text.strip()
        
        df = df.append({'Country': country, 'Gold': gold, 'Silver': silver, 'Bronze': bronze, 'Total': total}, ignore_index=True)
```

#### Exercise 1.2: Visualization
Create a visualization of the number of Gold, Silver, and Bronze medals for the top 5 countries with the most total medals using pandas

```python
# Visualization here
sorted_df_descending = df.sort_values(by='Total', ascending=False)
print(sorted_df_descending)
```
